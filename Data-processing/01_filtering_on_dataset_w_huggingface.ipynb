### 2-1. laion-1m-vit-h-14

```python
# load iterable dataset for conditional filtering
iterable_dataset = dataset.to_iterable_dataset()
```

```python
# find all the captions contain keyword => 'food'
import time
start_time = time.time()
count = 0
caption_index = []
captions = []
for item in iterable_dataset:
  caption = item['caption']
  if 'food' in caption.lower():
    caption_index.append(count)
    captions.append(caption)
    # print(caption)
    # break
  count += 1
  # if count > 1000000:
  #   break
end_time = time.time()
exec_time = end_time - start_time
```

### 2-2. ms-coco

```python
from datasets import load_dataset

dataset_coco = load_dataset("ChristophSchuhmann/MS_COCO_2017_URL_TEXT", split="train")
dataset_coco  # check dataset information
```

```python
import time
keywords_to_match = ['food', 'recipe', 'cook', 'restaurant',
                     'menu', 'delicious', 'cuisine', 'chef', 'dish', 'meal',
                     'plate','eat', 'bar','drink','plating','dip','cut','fork'
                     'breakfast','lunch','dinner','snack',
                     'barbecue','picnic',
                     'vegetable','bread','soup', 'egg','dessert','seafood','fried',
                     'culinary','ingredients','kitchen','gourmet','groceries',
                     'baker','homemade','foodie','post','taste','flavor','spice',
                     'seasoning','appetizer','main course','beverage',
                     'wine','coffee','coke','milk','juice','tea'
                     'street food','festival','fast food','vegetarian','vegan','muslim'
                     'organic','healthy','traval',
                     'delicious','tasty','bite','buffet',
                     'cut','chop','roast','grill','steam','stir','boil','toss',
                     'mix','slice','dice','peel','marinate','brown','melt','pour',
                     'nourish','nutrition','crave'
                     ]
def filter_food_coco(item):
  caption = item['TEXT']
  for keyword in keywords_to_match:
    if keyword.lower() in caption.lower():
      return True
  return False

start_time = time.time()
filtered_dataset_coco = dataset_coco.filter(filter_food_coco)
end_time = time.time()
exec_time = end_time - start_time
print(f"filter takes {exec_time} seconds")
```

```python
### multiple keyword searching
start_time = time.time()
count = 0
caption_index = []
matching_items = []
# List of keywords to match
keywords_to_match = ['food', 'recipe', 'cook', 'restaurant',
                     'menu', 'delicious', 'cuisine', 'chef', 'dish', 'meal',
                     'plate','eat', 'bar','drink','plating','dip','cut','fork'
                     'breakfast','lunch','dinner','snack',
                     'barbecue','picnic',
                     'vegetable','bread','soup', 'egg','dessert','seafood','fried',
                     'culinary','ingredients','kitchen','gourmet','groceries',
                     'baker','homemade','foodie','post','taste','flavor','spice',
                     'seasoning','appetizer','main course','beverage',
                     'wine','coffee','coke','milk','juice','tea'
                     'street food','festival','fast food','vegetarian','vegan','muslim'
                     'organic','healthy','traval',
                     'delicious','tasty','bite','buffet',
                     'cut','chop','roast','grill','steam','stir','boil','toss',
                     'mix','slice','dice','peel','marinate','brown','melt','pour',
                     'nourish','nutrition','crave'
                     ]

# Loop through the items in response['rows']
for item in iterable_dataset:
  caption = item['TEXT']
  for keyword in keywords_to_match:
    if keyword.lower() in caption.lower():
      caption_index.append(count)
      matching_items.append(item)
      break  # Break the loop after the first match to avoid duplicates
end_time = time.time()
exec_time = end_time - start_time
print(f"find {len(caption_index)} results related to {len(keywords_to_match)} keyword using {exec_time} seconds")
```
